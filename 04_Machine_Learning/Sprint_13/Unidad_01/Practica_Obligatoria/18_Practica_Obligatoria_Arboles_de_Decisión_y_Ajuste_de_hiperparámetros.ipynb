AYUDA: La direccin del banco nos proporciona estas notas (en ingls):

 
   **bank client data:**  
   1 - age (numeric)  
   2 - job : type of job (categorical: "admin.","unknown","unemployed","management","housemaid","entrepreneur","student",
                                       "blue-collar","self-employed","retired","technician","services")

  13 - campaign: number of contacts performed during this campaign and for this client (numeric, includes last contact)

```python
# Funcion para separar columnas y asi poder saber cules son las categoricas

def separar_columnas(df):
    # Inicializa listas para almacenar nombres de columnas categricas y numricas
    columnas_categoricas = []
    columnas_numericas = []

    # Itera sobre cada columna del DataFrame
    for columna in df.columns:
        # Si el tipo de datos de la columna es 'object', se considera categrica
        if df[columna].dtype == 'object':
            columnas_categoricas.append(columna)
        # Si el tipo de datos de la columna es numrico, se considera numrica
        else:
            columnas_numericas.append(columna)

    return columnas_categoricas, columnas_numericas

# Ejemplo de uso
# Suponiendo que tienes un DataFrame llamado 'df'
# Reemplaza 'df' con el nombre de tu DataFrame
columnas_categoricas, columnas_numericas = separar_columnas(df)
print("Columnas Categricas:", columnas_categoricas)
print("Columnas Numricas:", columnas_numericas)
```
