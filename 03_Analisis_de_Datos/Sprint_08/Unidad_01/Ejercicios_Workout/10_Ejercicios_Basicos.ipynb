### Ejercicio 0

Importa los paquetes y módulos que necesites a lo largo del notebook

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-white")
import numpy as np

x = np.linspace (0,10,100)
```

### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
y1 = np.sin(x)
y2 = np.cos(x)


plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primer subplot
plt.plot(x, y1)
plt.title('Seno')

plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segundo subplot
plt.plot(x, y2)
plt.title('Coseno')

plt.tight_layout()
plt.show()
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig, axs = plt.subplots(3, 1)  # 3 filas, 1 columna

axs[0].plot(x, y1, color = "red")  # Rojo
axs[1].plot(x, y2, color = "green")  # Verde
axs[2].plot(x, y3, color = "blue")  # Azul

plt.tight_layout()
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Línea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x, np.sin(2 * x), linestyle = "--", color= "magenta")  # Magenta, línea discontinua 'm--'
ax2.plot(x, np.cos(2 * x), linestyle = "-.", color = "cyan")  # Cian, línea-punto 'c-.'

plt.tight_layout()
plt.show()
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperbólico y coseno hiperbólico se obtienen así:
* np.sinh, seno hiperbólico;
* np.cosh, coseno hiperbólico;


```python
y1 = np.sinh(x)
y2 = np.cosh(x)

fig, axs = plt.subplots(2, 1)  # 2 filas, 1 columna


axs[0].plot(x, y1, color = "red")  # Rojo
axs[0].set_title ="Seno Hiperbólico"
axs[0].set_xlabel('x')
axs[0].set_ylabel('sinh(x)')

axs[1].plot(x, y2, color = "blue") #Azul
axs[1].set_title="Coseno Hiperbólico" 
axs[1].set_xlabel('x')
axs[1].set_ylabel('cosh(x)')

plt.tight_layout()
plt.show()
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.arctan(x)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, y3)
axs[1, 1].plot(x, y4)

fig.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```


### Ejercicio 6: 

Reproduce la figura siguiente empleando sólo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros gráficos (fila 1). Luego usa el estilo con métodos (objetos `axes`) para añadir el tercer gráfico.

```python
#Con MATLAB
plt.figure(figsize=(9,5)) # Esta línea crea una nueva figura con un tamaño específico de 9 pulgadas de ancho y 5 pulgadas de alto.
plt.subplot(2, 2, 1) #Esta línea indica que se creará una cuadrícula de subparcelas de 2x2 y que el siguiente gráfico se ubicará en la primera posición de esa cuadrícula.
plt.plot(x, np.cos(x), label = "coseno") #Aquí se traza la función coseno en función de x, que debe estar definida previamente, y se le asigna la etiqueta "coseno" para la leyenda


plt.subplot(2, 2, 1) #Esta línea vuelve a definir la primera subparcela, lo cual es redundante ya que la primera subparcela ya ha sido definida anteriormente. Podría ser un error tipográfico o un error de edición.
plt.plot(x, np.sqrt(x), label = "raiz cuadrada") #Aquí se traza la función raíz cuadrada en función de x, y se le asigna la etiqueta "raiz cuadrada" para la leyenda.

plt.legend() #Esta línea agrega una leyenda al gráfico para mostrar las etiquetas definidas anteriormente.
plt.title('Raíz Cuadrada y coseno') #Esta línea agrega un título al gráfico.


# Acceso a Axes para modificar detalles
ax = plt.subplot(2, 2, 3) # Esta línea indica que se creará un gráfico en la tercera posición de la cuadrícula de subparcelas y almacena el objeto de eje (Axes) en la variable ax
ax.plot(x, np.sin(x)) # Aquí se traza la función seno en función de x en el eje ax
ax.set_title('Seno') #Esta línea agrega un título al gráfico ubicado en el eje ax

plt.tight_layout() # Esta línea ajusta automáticamente el diseño del gráfico para que los elementos estén bien distribuidos y no se superpongan.
plt.show() # Esta línea muestra el gráfico completo.
```
