### Ejercicio 0

Importa los paquetes y mdulos que necesites a lo largo del notebook

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-white")
import numpy as np

x = np.linspace (0,10,100)
```

### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
y1 = np.sin(x)
y2 = np.cos(x)


plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primer subplot
plt.plot(x, y1)
plt.title('Seno')

plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segundo subplot
plt.plot(x, y2)
plt.title('Coseno')

plt.tight_layout()
plt.show()
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

fig, axs = plt.subplots(3, 1)  # 3 filas, 1 columna

axs[0].plot(x, y1, color = "red")  # Rojo
axs[1].plot(x, y2, color = "green")  # Verde
axs[2].plot(x, y3, color = "blue")  # Azul

plt.tight_layout()
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Lnea  

Modifica la primera figura para que se muestre de la siguiente manera

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x, np.sin(2 * x), linestyle = "--", color= "magenta")  # Magenta, lnea discontinua 'm--'
ax2.plot(x, np.cos(2 * x), linestyle = "-.", color = "cyan")  # Cian, lnea-punto 'c-.'

plt.tight_layout()
plt.show()
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperblico y coseno hiperblico se obtienen as:
* np.sinh, seno hiperblico;
* np.cosh, coseno hiperblico;


```python
y1 = np.sinh(x)
y2 = np.cosh(x)

fig, axs = plt.subplots(2, 1)  # 2 filas, 1 columna


axs[0].plot(x, y1, color = "red")  # Rojo
axs[0].set_title ="Seno Hiperblico"
axs[0].set_xlabel('x')
axs[0].set_ylabel('sinh(x)')

axs[1].plot(x, y2, color = "blue") #Azul
axs[1].set_title="Coseno Hiperblico" 
axs[1].set_xlabel('x')
axs[1].set_ylabel('cosh(x)')

plt.tight_layout()
plt.show()
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosicin (tal y como ves en la figura), emplea `subplots_adjust` y los parmetros hspace y wspace.

```python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.arctan(x)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, y3)
axs[1, 1].plot(x, y4)

fig.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```


### Ejercicio 6: 

Reproduce la figura siguiente empleando slo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros grficos (fila 1). Luego usa el estilo con mtodos (objetos `axes`) para aadir el tercer grfico.

```python
#Con MATLAB
plt.figure(figsize=(9,5)) # Esta lnea crea una nueva figura con un tamao especfico de 9 pulgadas de ancho y 5 pulgadas de alto.
plt.subplot(2, 2, 1) #Esta lnea indica que se crear una cuadrcula de subparcelas de 2x2 y que el siguiente grfico se ubicar en la primera posicin de esa cuadrcula.
plt.plot(x, np.cos(x), label = "coseno") #Aqu se traza la funcin coseno en funcin de x, que debe estar definida previamente, y se le asigna la etiqueta "coseno" para la leyenda


plt.subplot(2, 2, 1) #Esta lnea vuelve a definir la primera subparcela, lo cual es redundante ya que la primera subparcela ya ha sido definida anteriormente. Podra ser un error tipogrfico o un error de edicin.
plt.plot(x, np.sqrt(x), label = "raiz cuadrada") #Aqu se traza la funcin raz cuadrada en funcin de x, y se le asigna la etiqueta "raiz cuadrada" para la leyenda.

plt.legend() #Esta lnea agrega una leyenda al grfico para mostrar las etiquetas definidas anteriormente.
plt.title('Raz Cuadrada y coseno') #Esta lnea agrega un ttulo al grfico.


# Acceso a Axes para modificar detalles
ax = plt.subplot(2, 2, 3) # Esta lnea indica que se crear un grfico en la tercera posicin de la cuadrcula de subparcelas y almacena el objeto de eje (Axes) en la variable ax
ax.plot(x, np.sin(x)) # Aqu se traza la funcin seno en funcin de x en el eje ax
ax.set_title('Seno') #Esta lnea agrega un ttulo al grfico ubicado en el eje ax

plt.tight_layout() # Esta lnea ajusta automticamente el diseo del grfico para que los elementos estn bien distribuidos y no se superpongan.
plt.show() # Esta lnea muestra el grfico completo.
```
