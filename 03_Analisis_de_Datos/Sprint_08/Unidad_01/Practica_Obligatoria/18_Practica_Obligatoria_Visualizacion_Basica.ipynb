### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
'''1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center").
(Colores: azul, gris, aceituna, marrn, con trasparencia)'''
```

```python
#sacar las mas pobladas
mas_pobladas = df_california.loc[df_california.population_total.nlargest(5).index, ["city","population_total"]]

#modificar los nombres y remplazarlos
modificadas = {"LosAngeles": "Los Angeles", "SanDiego": "San Diego", "SantaAna": "Santa Ana"}
ciudades = mas_pobladas["city"].replace (modificadas).to_list()

#crear la figura
plt.figure(figsize = (10,5))
plt.bar(x= ciudades, height= mas_pobladas["population_total"]/1000, color = ("blue","gray","red","olive","brown"), alpha = 0.5)
plt.ylim([0,5000])
plt.ylabel("Habitantes (miles)")
plt.xlabel("Ciudades")


mas_pobladas.set_index("city", inplace = True)
for indice, ciudad in enumerate(mas_pobladas.index):
    poblacion = mas_pobladas.loc[ciudad,"population_total"]/1000
    plt.text(ciudades[indice], poblacion + 50, f"{poblacion}", ha = "center")
    
plt.title("Ciudades ms pobladas de California")
plt.show()
```

```python
'''2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)'''
```

```python
plt.figure (figsize = (10,5))
plt.hist(df_california["area_total_km2"], color = "steelblue", bins = 50, label= "Hist Area")
plt.legend()
plt.ylabel("Nmero de ciudades", fontdict=dict(family ="Bradley Hand itc"), size = 18)
plt.xlabel("Superficie($km^2$)", fontdict=dict(family="Bradley Hand itc"), size = 18)
plt.show()
```

```python
'''3. El scatter de rea de agua de un ciudad con su poblacion.'''
```

```python
plt.figure (figsize=(10,5))
plt.scatter (df_california["area_water_km2"], df_california["population_total"]/1000, color="green", alpha = 0.5)
plt.ylabel("Poblacin (miles de habitantes)")
plt.xlabel("rea de agua en la ciudad ($km^2$)")
plt.title("Poblacin por kilmetro cuadrado de agua")

# Identificacin y sealizacin de outliers (HE USADO IA, porque no saba)
outlier_1_y = df_california["population_total"].max()
outlier_2_y = df_california["population_total"].nlargest(2).iloc[1]  # El segundo valor ms grande
outlier_1_x = df_california.loc[df_california["population_total"] == outlier_1_y, "area_water_km2"].values[0]
outlier_2_x = df_california.loc[df_california["population_total"] == outlier_2_y, "area_water_km2"].values[0]

plt.annotate("Outliers", xy=(outlier_1_x, outlier_1_y/1000), xytext=(60, 3500),
             arrowprops=dict(arrowstyle="->"))
plt.annotate("Outliers", xy=(outlier_2_x, outlier_2_y/1000), xytext=(60, 3500),
             arrowprops=dict(arrowstyle="->"))

plt.show()
```

```python
'''4. El boxplot de la altura de las ciudades.'''
```

```python
plt.figure (figsize=(10,5))
plt.boxplot(df_california["elevation_m"],vert=False, patch_artist=True, boxprops=dict(facecolor="red"), whiskerprops=dict(color="black"), medianprops=dict(color="black"))
plt.xlabel("Altitud(m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
```

```python
# Juntar todas

fig, ax = plt.subplots (2,2, figsize = (20,20))

#primera grafica
mas_pobladas = df_california.loc[df_california.population_total.nlargest(5).index, ["city","population_total"]]
modificadas = {"LosAngeles": "Los Angeles", "SanDiego": "San Diego", "SantaAna": "Santa Ana"}
ciudades = mas_pobladas["city"].replace (modificadas).to_list()

ax[0,0].bar(x= ciudades, height= mas_pobladas["population_total"]/1000, color = ("blue","gray","red","olive","brown"), alpha = 0.5)
ax[0,0].set_ylim([0,5000])
ax[0,0].set_ylabel("Habitantes (miles)")
ax[0,0].set_xlabel("Ciudades")

mas_pobladas.set_index("city", inplace = True)
for indice, ciudad in enumerate(mas_pobladas.index):
    poblacion = mas_pobladas.loc[ciudad,"population_total"]/1000
    ax[0,0].text(ciudades[indice], poblacion + 50, f"{poblacion}", ha = "center")
    
ax[0,0].set_title("Ciudades ms pobladas de California")


#segunda
ax[0,1].hist(df_california["area_total_km2"], color = "steelblue", bins = 50, label= "Hist Area", alpha = 0.7)
ax[0,1].legend()
ax[0,1].set_ylabel("Nmero de ciudades", fontdict=dict(family ="Bradley Hand itc"), size = 18)
ax[0,1].set_xlabel("Superficie($km^2$)", fontdict=dict(family="Bradley Hand itc"), size = 18)

#tercera
ax[1,0].scatter (df_california["area_water_km2"], df_california["population_total"]/1000, color="green", alpha = 0.5)
ax[1,0].set_ylabel("Poblacin (miles de habitantes)")
ax[1,0].set_xlabel("rea de agua en la ciudad ($km^2$)")
ax[1,0].set_title("Poblacin por kilmetro cuadrado de agua")

outlier_1_y = df_california["population_total"].max()
outlier_2_y = df_california["population_total"].nlargest(2).iloc[1] 
outlier_1_x = df_california.loc[df_california["population_total"] == outlier_1_y, "area_water_km2"].values[0]
outlier_2_x = df_california.loc[df_california["population_total"] == outlier_2_y, "area_water_km2"].values[0]

ax[1,0].annotate("Outliers", xy=(outlier_1_x, outlier_1_y/1000), xytext=(60, 3500),
             arrowprops=dict(arrowstyle="->"))
ax[1,0].annotate("Outliers", xy=(outlier_2_x, outlier_2_y/1000), xytext=(60, 3500),
             arrowprops=dict(arrowstyle="->"))

#cuarta
ax[1,1].boxplot(df_california["elevation_m"],vert=False, patch_artist=True, boxprops=dict(facecolor="red"), whiskerprops=dict(color="black"), medianprops=dict(color="black"))
ax[1,1].set_xlabel("Altitud(m)")
ax[1,1].set_title("Diagrama de caja de la Altitud de las ciudades de California")
;
```
